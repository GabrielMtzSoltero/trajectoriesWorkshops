```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

#Creamos una imagen negra
cuadro = np.zeros((100, 100, 3), dtype=np.uint8)

#accedemos a una parte de la imagen anterior para ponerle negro
cuadro[20:80, 20:80] = [255, 255, 255]

#modificamos un canal
#cuadro[:, :, 1] = 0 # Eliminamos el rojo del canal 2 (BGR)

print("Imagen generada puramente con NumPy:")
cv2_imshow(cuadro)

```

```python
print("\n--- Leyendo imagen desde archivo ---")

# Reemplaza 'path/to/your/image.jpg' con la ruta a tu archivo de imagen
# Asegúrate de que el archivo exista en la ruta especificada
image_path = 'path/to/your/image.jpg'

# Leemos la imagen
# cv2.imread devuelve None si el archivo no se encuentra o no se puede leer
imagen_leida = cv2.imread(image_path)

# Verificamos si la imagen se cargó correctamente
if imagen_leida is not None:
    print(f"Imagen '{image_path}' leída correctamente:")
    cv2_imshow(imagen_leida)
else:
    print(f"Error: No se pudo leer la imagen en la ruta '{image_path}'.")
    print("Por favor, verifica la ruta y el nombre del archivo.")
```
